```python
import numpy as np
import pandas as pd
```

```python
product_name = np.random.choice(['Moto G6', 'Apple iPhone 6', 'Honor 7X', 'BLU Studio X8', 'Moto E (4th Generation)', 'Moto G PLUS', 'Nokia 6.1', 'Lg Vn170', 'Moto X (4th Generation)', 'Apple iPhone', 'Samsung Galaxy S9+'], 1000, p=[0.01, 0.09, 0.04, 0.06, 0.2, 0.1 ,0.1, 0.04, 0.06, 0.2, 0.1])
shipping_type = np.random.choice(['Free', 'Customer Paid'], 1000, p=[0.7,0.3])
customer_type = np.random.choice(['New', 'Existing'], 1000, p=[0.2,0.8])
location = np.random.choice(['Sydney', 'Newcastle', 'Central Coast', 'Wollongong', 'Albury'], 1000, p=[0.2,0.5,0.1,0.1,0.1])
```

```python
dataset = pd.DataFrame({'Product_name' : product_name, 'Shipping_Type' : shipping_type, 'Customer_type' : customer_type, 'Location' : location})
dataset.head()
```

```python
product_data = pd.DataFrame(columns = ['Product_name', 'Product_price', 'Monthly_sales', 'No_of_cust_who_bought'])
product_data['Product_name'] = ['Moto G6', 'Apple iPhone 6', 'Honor 7X', 'BLU Studio X8', 'Moto E (4th Generation)', 'Moto G PLUS', 'Nokia 6.1', 'Lg Vn170', 'Moto X (4th Generation)', 'Apple iPhone', 'Samsung Galaxy S9+']
product_data['Product_price'] = [100, 245, 205, 97, 90, 120, 300, 200, 180, 101, 100]
product_data['Monthly_sales'] = [1571, 4080, 280, 3004, 7527, 1000, 705, 1789, 4000, 1560, 5571]
product_data
```

```python
dataset = dataset.merge(product_data, right_on = 'Product_name', left_on = 'Product_name', how = 'outer')
```
